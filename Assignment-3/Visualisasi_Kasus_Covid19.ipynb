```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Dataset-Kasus-COVID19-shortened.csv")
data.head()
```

```python
data.info()
```

```python
data['Nama Negara'].unique()
```

```python
unique_countries = data[data['Total Terinfeksi']>0]['Nama Negara'].unique()
unique_countries.sort()
unique_countries
```

```python
country_confirmed_case = []
for i in unique_countries:
  country_confirmed_case.append(data[data['Total Terinfeksi']>0][data['Nama Negara']==i]['Total Terinfeksi'].sum())
  
```

```python
print(set(zip(unique_countries, country_confirmed_case)))
```

```python
nan_indices = []
for i in range(len(unique_countries)):
  if type(unique_countries[i])==float:
    nan_indices.append(i)
```

```python
unique_countries = list(unique_countries)
country_confirmed_case = list(country_confirmed_case)
```

```python
for i in nan_indices:
  unique_countries.pop(i)
  country_confirmed_case.pop(i)

unique_countries
```

```python
plt.barh(unique_countries, country_confirmed_case)
plt.show()
```

```python
plt.figure(figsize=(50,40))
plt.barh(unique_countries, country_confirmed_case)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(country_confirmed_case)
plt.legend(unique_countries, loc='best')
plt.show()
```
